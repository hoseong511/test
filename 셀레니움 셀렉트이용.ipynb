```python
import pandas as pd
import bs4
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait  # 페이지가 로딩될때까지 기다리는 기능
from selenium.webdriver.support import expected_conditions as EC  # 페이지 로딩 조건 기능
from selenium.webdriver.common.by import By

driver_loc="./chromedriver/chromedriver.exe" # 드라이버 위치

options=webdriver.ChromeOptions().add_argument('window-size=2560x1440') # 옵션
driver=webdriver.Chrome(driver_loc, options=options) # 드라이버 생성

driver.implicitly_wait(3) # 페이지가 로딩될때까지 3초까지 기다린다.

base_url='https://sparkkorea.com/' # url
driver.get(base_url) # url로 이동

quizBtn = '//*[@id="menu-item-382"]/a'
driver.find_element_by_xpath(quizBtn).click() # 퀴즈 버튼을 xpath 경로로 찾아서 클릭

WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.ID, 'id_spark_quiz'))) # id가 id_spark_quiz 인 요소가 확인될때까지 기다립니다.

html=driver.page_source # 페이지 소스 받아오기
bs = bs4.BeautifulSoup(html, "html.parser") # 파싱

quizData = bs.body.select("div#id_spark_quiz.class_spark_quiz > a[href]")
#   id가 id_spark_quiz 이고 class가 class_spark_quiz 인 div태그를 찾음
# > div 태그의 자식이면서 href 속성을 가진 a태그를 모두 select

row=[]
dataFrame = []
# 행을 담을 리스트, 전체를 담을 리스트

for data in quizData:
    row.append(data.get_text(strip=True))
    row.append(data['href'])
    #행에 링크, 타이틀을 담음
    
    dataFrame.append(row)
    #전체 리스트에 행 리스트를 추가
    
    row=[]
    #행 리스트는 초기화
    
result = pd.DataFrame(dataFrame, columns =["spark퀴즈 타이틀", "spark퀴즈 링크"])
#데이터프레임

result
```
