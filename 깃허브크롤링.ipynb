quit()는 dispose() 함수를 불러와 열려있는 모든 webdriver를 종료하고 세션을 안전하게 종료합니다.\
프로그램을 종료할 때 quit()을 사용하지 않는다면 webdriver 세션이 완벽하게 종료되지 않아 메모리 누수가 발생할 수 있습니다.

```python
import pandas as pd
import psycopg2
import pymysql
from sqlalchemy import create_engine

def sqlReader(query, URL):
    engine=create_engine(URL)
    dataFrame=pd.read_sql_query(query, engine)
    engine.dispose()
    return dataFrame
#query, URL 을 입력하면 데이터프레임으로 돌려주는 함수
#

postgres_URL = "postgresql+psycopg2://haiteam:haiteam@www.hadoopkorea.com:5432/postgres"
mariadb_URL = "mysql+pymysql://kopo:kopo@www.hadoopkorea.com:3306/kopo"
#접속 URL

postgres_query = "SELECT * FROM kopo_product_volume_st01"
mariadb_query = "SELECT * FROM kopo_product_volume_st02"
#쿼리문 입력

pDataFrame = sqlReader(postgres_query, postgres_URL)
mDataFrame = sqlReader(mariadb_query, mariadb_URL)
#데이터프레임으로 불러오기

mDataFrame.columns = mDataFrame.columns.str.lower()

answer = pd.concat([pDataFrame, mDataFrame], axis = 0)
# 하나로 합치기

answer


```
