```python
import re
# 키워드리스트에는 한 가지 성향으로만!! -> 긍정적인 표현만 묶기

# 속도 키워드
key1List = ['바','ab','abf','ab','asbf','ba','badf','ba']
# 무게 키워드
key2List = ['ab','ab','abf','ab','asbf','가','badf','ba']
# 가성비 키워드
key3List = ['ab','ab','abf','ab','asbf','ba','badf','ba']
# 디자인 키워드
key4List = ['ab','ab','abf','ab','asbf','ba','badf','ba']
# 댓글
commentList = ['가나다라마바사자차','b','c','d','e','f']
velocityPostive = 0
weightPostive = 0
designPostive = 0
benefitPostive = 0

# 전체 댓글들이 담긴 리스트만큼 반복문을 전부 돌림
for i in range(len(commentList)):
#     1차 정제 -> 조건을 맞지 않는 않는 것들은 버림
    comment = commentList[i]
    print(comment)
    if (None == re.search('배송', comment)) & (len(commentList) > 5):
        
        
        for j in range(len(key1List)):
            
            keyWord = key1List[j]
            print( keyWord, i)
#             댓글에 속도 긍정 표현있다면 +1
            if None != re.search(keyWord, comment):
                velocityPostive += 1
            
        for j in range(len(key2List)):
            keyWord = key2List[j]
#             무게 긍정 표현있다면 +1
            if None != re.search(keyWord, comment):
                weightPostive += 1
            
        for j in range(len(key3List)):
            keyWord = key3List[j]
#             가성비 긍정 표현있다면 +1
            if None != re.search(keyWord, comment):
                benefitPostive += 1
            
        for j in range(len(key4List)):
            keyWord = key4List[j]
#             디자인 긍정 표현 있다면 +1
            if None != re.search(keyWord, comment):
        
                designPostive += 1
                
        
    else:
        continue
        
print("속도plus",velocityPostive)
print("가성비plus",benefitPostive)
print("무게plus",weightPostive)
print("디자인plus",designPostive)
```
