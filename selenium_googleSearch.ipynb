```python
# <스마트금융과 네이버 카페에서 크롤링하기>

import bs4,requests,time,pyperclip
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait  # 페이지가 로딩될때까지 기다리는 기능
from selenium.webdriver.support import expected_conditions as EC  # 페이지 로딩 조건 기능
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver_loc = "C:/Users/Lenovo/pywork/chromedriver/chromedriver.exe"
options = webdriver.ChromeOptions()
options.add_argument("window-size = 1920x1080")
driver = webdriver.Chrome(driver_loc, options=options)

naverUrl = 'https://www.naver.com'
cafeUrl = 'https://cafe.naver.com/smartpolytech'

driver.get(naverUrl)
driver.maximize_window() 
time.sleep(2)

inputId = ''
inputPw = ''
inputSearch = '채용'
login_btn = driver.find_element_by_class_name('link_login')
login_btn.click()
time.sleep(1)

# id, pw 입력할 곳을 찾습니다. -> 태크 중 id속성의 이름을 각각 찾아낸다
tag_id = driver.find_element_by_id('id')
tag_pw = driver.find_element_by_id('pw')
time.sleep(1)

# id 입력 -> (네이버)id, pw 입력할 때 sendkey에 직접 값을 넣게되면 자동입력방지가 실행됨
tag_id.click()
pyperclip.copy(inputId)
tag_id.send_keys(Keys.CONTROL, 'v')
time.sleep(1)

# pw 입력 
tag_pw.click() 
pyperclip.copy(inputPw) 
tag_pw.send_keys(Keys.CONTROL, 'v') 
time.sleep(1) 

# 로그인 버튼을 클릭합니다 
login_btn = driver.find_element_by_id('log.login') 
login_btn.click()
time.sleep(1)

# 카페 접속
driver.get(cafeUrl)

# 카페 내 검색창 찾기
tag_search = driver.find_element_by_name('query')

# 검색창 클릭
tag_search.click()
pyperclip.copy(inputSearch)
tag_search.send_keys(Keys.CONTROL, 'v')
time.sleep(1)
driver.find_element_by_xpath('//*[@id="info-search"]/form/button').click()

time.sleep(2)
iframes = driver.find_elements_by_css_selector('iframe')
    
driver.switch_to_default_content 
driver.switch_to.frame("cafe_main")
time.sleep(1)

data = driver.page_source
bs = bs4.BeautifulSoup(data, 'html.parser')

# 태그 이름이 div이고 속성 class의 이름이 article-board result-board m-tcol-c인것 찾기
divTag = bs.find(name = 'div',
                attrs = {'class': 'article-board result-board m-tcol-c'})
trTags = divTag.find(name = 'tbody').findAll(name = 'tr')
rowList =[]

for i in range(0, len(trTags)):
    columnList = []
    # 홀수번째 자료는 제외 -> 필요없는 내용
    if (i % 2 != 0):
        continue
    each_columns = trTags[i].findAll(name = 'td')
    
    for j in range(0,len(each_columns)):
        # 2번째 자료 제외 -> 필요없는 내용
        if j == 1:
            continue
        columnList.append(each_columns[j].text.replace(" ","").replace("\n",""))
    rowList.append(columnList)    
result = pd.DataFrame(rowList, columns = ["제목", "작성자", "작성일", "조회수"])
result
# result.to_csv('./navercafe.csv', encoding= 'ms949', index = False)
```
