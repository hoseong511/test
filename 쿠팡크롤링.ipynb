```python
import pandas as pd
import bs4
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',}
# 헤더 정의

dataFrameList=[]
# 모든 결과를 담을 리스트 생성

for i in range(0, 17):   # 페이지가 17개 있어서 17번 반복합니다.
    url = 'https://www.coupang.com/np/categories/194810?page='+str(i+1)  # URL 뒷자리만 바꾸면서 17페이지까지 이동
    data=requests.get(url,headers=headers) # URL 에서 리스폰스를 받아옵니다
    data.encoding='utf-8'  # 인코딩
    html = data.text       # 원본 소스
    bs = bs4.BeautifulSoup(html, "html.parser")  # BeautifulSoup 로 예쁘게 파싱
    
    products = bs.select("a.baby-product-link")  # 변수 products 는 select로 골라왔으므로 리스트형식입니다.
    columns = []    # 행을 담기위한 리스트

    for product in products:  # products 리스트의 모든 요소들에 대해서
        name = product.select_one("dd.descriptions").select_one(".name").get_text(strip=True)  # 상품명 부분입니다.
        price = product.select_one("dd.descriptions").select_one(".price-value").get_text(strip=True)  # 가격 부분입니다.
        columns.append(name)   # 행에 상품명을 담습니다.
        columns.append(price)  # 행에 가격을 담습니다.
        dataFrameList.append(columns) # 최종결과 리스트에 행을 추가합니다.
        columns = []           # 행은 초기화해줍니다.
        
    




result = pd.DataFrame(dataFrameList, columns=['상품명', '가격']) 
# 최종결과를 dataframe으로 포장합니다.

result
```

```python
# 라이브러리 불러오기
import requests,bs4
import pandas as pd


# 웹페이지 요청 후 가져옴
url = 'http://www.ssg.com/search.ssg?target=all&query=%EA%B3%84%EB%9E%80'
data = requests.get(url)

# data를 인코딩하기
data.encoding = 'utf-8'

# 텍스트화하기
html = data.text

# bs4가 html형식으로 파싱
bs = bs4.BeautifulSoup(html,'html.parser')

# 상품 정보를 포괄하는 태그 div에 class속성이 tmpl_itemlist를 찾기
div_tags = bs.find(name = 'div',
                 attrs = {'class': 'tmpl_itemlist'})

# 태그 div에 class 속성 cunit_info를 모두 찾기
div_info_tags = div_tags.findAll(name = 'div',
                    attrs = {'class':'cunit_info'})

# 빈 리스트 저장 -> 상품명과 가격을 담을 리스트
totalList =[]
div_info_tagsLen = len(div_info_tags)

# 반복문으로 리스트에 담기
for i in range (0, div_info_tagsLen):
    
    # eachName과 eachPrice를 태그 em에 각각 clss속성에 tx_ko, ssg_price
    eachName = div_info_tags[i].find(name = 'em',
                           attrs = {'class':'tx_ko'})
    
    eachPrice = div_info_tags[i].find(name = 'em',
                               attrs = {'class':'ssg_price'})
    
    # 텍스트화하기
    eachName = eachName.text
    eachPrice = eachPrice.text+"원"
    
    # 리스트에 추가하기
    totalList.append([eachName, eachPrice])
    
# 컬럼명이 상품명과 가격인 데이터프레임으로 저장
result = pd.DataFrame(totalList, columns=["상품명","가격"])
result
```
